Удаление шипов
Шипы - это линии, отмечающие границы данных и соединяющие метки оси. Его можно удалить с помощью метода despine().

```python

# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("iris")
 
# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
 
# Removing the spines
sns.despine()
plt.show()
```

Изменение размера фигуры
Размер рисунка можно изменить с помощью метода figure() Matplotlib. метод figure() создает новую фигуру указанного размера, переданную в параметре figsize.

 

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("iris")
 
# changing the figure size
plt.figure(figsize = (2, 4))
 
# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
 
# Removing the spines
sns.despine()
 
plt.show()
```

Настройка цветовой палитры по умолчанию
метод set_palette() используется для установки цветовой палитры по умолчанию для всех графиков. Аргументы как для color_palette(), так и для set_palette() одинаковы. set_palette() изменяет параметры matplotlib по умолчанию.

 

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("iris")
 
def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)
 
# setting the default color palette
sns.set_palette('vlag')
plt.subplot(211)
 
# plotting with the color palette
# as vlag
plot()
 
# setting another default color palette
sns.set_palette('Accent')
plt.subplot(212)
plot()
 
plt.show()
```

Использование Matplotlib
Matplotlib предоставляет различные функции для построения вложенных графиков. Некоторые из них - add_axes(), subplot() и subplot2grid(). Давайте посмотрим пример каждой функции для лучшего понимания.

 

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
 
# loading dataset
data = sns.load_dataset("iris")
 
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)
 
# Creating a new figure with width = 5 inches
# and height = 4 inches
fig = plt.figure(figsize =(5, 4))
 
# Creating first axes for the figure
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
 
# plotting the graph
graph()
 
# Creating second axes for the figure
ax2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
 
# plotting the graph
graph()
 
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("iris")
 
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)
 
# adding the subplots
axes1 = plt.subplot2grid (
  (7, 1), (0, 0), rowspan = 2,  colspan = 1)
graph()
 
axes2 = plt.subplot2grid (
  (7, 1), (2, 0), rowspan = 2, colspan = 1)
graph()
   
axes3 = plt.subplot2grid (
  (7, 1), (4, 0), rowspan = 2, colspan = 1)
graph()
```

Метод 1: использование метода FacetGrid()
 

Класс FacetGrid помогает визуализировать распределение одной переменной, а также взаимосвязь между несколькими переменными отдельно в подмножествах вашего набора данных с использованием нескольких панелей.
Фасетную сетку можно нарисовать с точностью до трех измерений ? строка, цвет и оттенок. Первые два имеют очевидное соответствие с результирующим массивом осей; представьте переменную hue как третье измерение вдоль оси глубины, где разные уровни отображаются разными цветами.
Объект FacetGrid принимает фрейм данных в качестве входных данных и имена переменных, которые будут формировать размеры строк, столбцов или оттенков сетки. Переменные должны быть категориальными, и данные на каждом уровне переменной будут использоваться для фасета вдоль этой оси.
 

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("iris")
 
plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")
 
plt.show()
```

Вспомогательная сетка для построения попарных связей в наборе данных.

Он также может представлять дополнительный уровень условности с параметром hue, который отображает различные подмножества данных в разных цветах. Это использует цвет для разрешения элементов в третьем измерении, но только рисует подмножества друг над другом и не будет адаптировать параметр hue для конкретной визуализации так, как это делают функции уровня осей, которые принимают оттенок.

Реляционные графики
Реляционные графики используются для визуализации статистической взаимосвязи между точками данных. Визуализация необходима, потому что она позволяет человеку видеть тенденции и закономерности в данных. Процесс понимания того, как переменные в наборе данных соотносятся друг с другом, и их взаимосвязи называются статистическим анализом. Relplot()
Эта функция предоставляет нам доступ к некоторым другим различным функциям уровня axes, которые показывают отношения между двумя переменными с семантическими отображениями подмножеств. Он строится с использованием метода relplot().

Distplot используется в основном для одномерного набора наблюдений и визуализирует его с помощью гистограммы, то есть только одного наблюдения, и, следовательно, мы выбираем один конкретный столбец набора данных. Он заполняется с помощью метода distplot().

Совместная схема
Jointplot используется для построения графика двух переменных с двумерными и одномерными графами. Он в основном сочетает в себе два разных сюжета. Он строится с использованием метода jointplot().

Схема сопряжения
 

Pairplot представляет попарное отношение по всему фрейму данных и поддерживает дополнительный аргумент с именем hue для категориального разделения. По сути, он создает объединенную диаграмму между всеми возможными числовыми столбцами и занимает некоторое время, если фрейм данных действительно огромен. Он строится с использованием метода pairplot().

 

```python

# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("iris")
 
sns.pairplot(data=data, hue='species')
plt.show()
```

Коврик для ковра
 

Rugplot отображает точки данных в массиве в виде точек на оси.Так же, как и distplot, он занимает один столбец. Вместо того чтобы рисовать гистограмму, он создает прочерки по всему участку. Если вы сравните его с joinplot, то увидите, что jointplot делает то, что он подсчитывает тире и показывает их в виде ячеек. Он строится с использованием метода rugplot().

```python

# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("iris")
 
sns.rugplot(data=data)
plt.show()
```

График KDE
График KDE, описанный как оценка плотности ядра, используется для визуализации плотности вероятности непрерывной переменной. Он отображает плотность вероятности при разных значениях непрерывной переменной. Мы также можем построить один график для нескольких выборок, что помогает в более эффективной визуализации данных.

 

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("iris")
 
sns.kdeplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

Графики регрессии
Графики регрессии в первую очередь предназначены для добавления визуального руководства, которое помогает выделить закономерности в наборе данных во время предварительного анализа данных. Графики регрессии, как следует из названия, создают линию регрессии между двумя параметрами и помогают визуализировать их линейные взаимосвязи.

есть две основные функции, которые используются для построения моделей линейной регрессии. Эти функции lmplot() и regplot() тесно связаны друг с другом. Они даже разделяют свою основную функциональность.

 

lmplot
метод lmplot() можно понимать как функцию, которая в основном создает линейный график модели. Он создает точечный график с линейной подгонкой поверх него.

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("tips")
 
sns.lmplot(x='total_bill', y='tip', data=data)
plt.show()
```

Матричные графики
Матричный график означает построение матричных данных, в которых диаграммы с цветовой кодировкой отображают данные строк, данные столбцов и значения. Его можно отобразить с помощью тепловой карты и карты кластеров.

 

Тепловая карта
Тепловая карта определяется как графическое представление данных с использованием цветов для визуализации значения матрицы. В этом случае для представления более распространенных значений или более высоких действий используются более яркие цвета, в основном красноватые, а для представления менее распространенных значений или значений активности предпочтительны более темные цвета. его можно построить с помощью функции heatmap() .

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("tips")
 
# correlation between the different parameters
tc = data.corr()
 
sns.heatmap(tc)
plt.show()
 
```

Карта кластеров
Функция clustermap() seaborn отображает иерархически кластеризованную тепловую карту заданного матричного набора данных. Кластеризация просто означает группировку данных на основе взаимосвязи между переменными в данных.

 

```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt
 
# loading dataset
data = sns.load_dataset("tips")
 
# correlation between the different parameters
tc = data.corr()
 
sns.clustermap(tc)
plt.show()
```
